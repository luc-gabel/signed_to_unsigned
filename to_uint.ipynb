## Signed images to unsigned
A simple script to convert your signed tif images to unsigned, so you can open them in Krita.

```python
from skimage import io
import glob

from helpers import *
```

```python
img = io.imread('file.tif')

plt.figure(figsize=(8,8))
plot=plt.imshow(img, cmap="gray");
add_colorbar(plot)
plt.show()
```

This is a __16-bit signed integer image__, which means the values are in the range (-32'768, 32'767).

As you can see, the values go into the negatives, but Krita only supports unsigned (positive-valued) images. To solve this, we need to make sure that values are in the range (0, 65'535). In short, we want an __unsigned image__.

Fortunately, we can very easily code a function that does the conversion (see helpers.py).

```python
unsigned_img=to_uint(img) #for 16-bit

plt.figure(figsize=(8,8))
plot=plt.imshow(unsigned_img, cmap="gray");
add_colorbar(plot)
plt.show()
```
